```python
%%writefile language_translator.py
import streamlit as st
import torch
from transformers import pipeline
import warnings
import os

# Suppress warnings
os.environ['TOKENIZERS_PARALLELISM'] = 'false'
warnings.filterwarnings("ignore", category=FutureWarning)

st.set_page_config(page_title="Language Translator", page_icon="🌐", layout="wide")

# App Title
st.markdown('<h1 style="text-align:center; color:#1E88E5;">🌍 Language Translator</h1>', unsafe_allow_html=True)

# Language codes supported by Helsinki-NLP models
LANGUAGES = {
    "English": "en", "Spanish": "es", "French": "fr", "German": "de",
    "Italian": "it", "Portuguese": "pt", "Russian": "ru", "Chinese": "zh",
    "Japanese": "ja", "Arabic": "ar", "Hindi": "hi"
}

if "device" not in st.session_state:
    st.session_state.device = 0 if torch.cuda.is_available() else -1
if "models" not in st.session_state:
    st.session_state.models = {}

# Function to dynamically load the right model
@st.cache_resource
def load_model(source_lang, target_lang):
    model_name = f"Helsinki-NLP/opus-mt-{LANGUAGES[source_lang]}-{LANGUAGES[target_lang]}"
    return pipeline("translation", model=model_name, device=st.session_state.device)

def translate_text(text, src, tgt):
    if not text.strip():
        return ""
    try:
        key = f"{src}-{tgt}"
        if key not in st.session_state.models:
            st.session_state.models[key] = load_model(src, tgt)
        model = st.session_state.models[key]
        out = model(text, max_length=400)
        return out[0]["translation_text"]
    except Exception as e:
        return f"⚠️ Translation error: {str(e)}"

# Layout
col1, col2 = st.columns(2)

with col1:
    source_lang = st.selectbox("Source language:", list(LANGUAGES.keys()), index=0)
    source_text = st.text_area("Enter text:", height=200)

with col2:
    target_lang = st.selectbox("Target language:", [l for l in LANGUAGES if l != source_lang], index=1)
    output_area = st.empty()

if st.button("Translate"):
    result = translate_text(source_text, source_lang, target_lang)
    output_area.text_area("Translated:", value=result, height=200, disabled=True)
```
