```python
import os

# Base path to your dataset images
base_path = '/kaggle/input/intrusion-detection/Intrusion Detection/Dataset'

for split in ['train', 'valid', 'test']:
    path = os.path.join(base_path, split, 'images')
    if os.path.exists(path):
        count = len([f for f in os.listdir(path) if f.endswith(('.jpg', '.jpeg', '.png'))])
        print(f"📊 {split.capitalize()} images: {count}")
    else:
        print(f"⚠️ {split.capitalize()} folder not found at {path}")
```

```python
from ultralytics import YOLO

# Load a pretrained YOLOv8 Nano model
model = YOLO('yolov8n.pt')

# Train the model
model.train(
    data='/kaggle/input/intrusion-detection/Intrusion Detection/Dataset/data.yaml',
    epochs=50,
    imgsz=416,
    batch=16,
    device=0,      # Use GPU
    name='person_detection_model'
)
```

```python
# Load the best weights from your training run
best_model = YOLO('/kaggle/working/runs/detect/person_detection_model/weights/best.pt')

# Run formal validation
val_metrics = best_model.val(split='val') # Uses the validation set from data.yaml

print(f"--- VALIDATION METRICS ---")
print(f"mAP50: {val_metrics.box.map50:.4f}")
print(f"Precision: {val_metrics.box.mp:.4f}")
print(f"Recall: {val_metrics.box.mr:.4f}")
print(f"Inference Speed: {val_metrics.speed['inference']:.2f}ms per image")
```

```python
import time
from ultralytics import YOLO

# 1. Load your best trained weights
model = YOLO('/kaggle/working/runs/detect/person_detection_model/weights/best.pt')

# 2. Accuracy & Formal Metrics (on Test Set)
# We use split='test' to get official accuracy scores for the 9 images
print("\n--- [ACCURACY REPORT] ---")
test_metrics = model.val(
    data='/kaggle/input/intrusion-detection/Intrusion Detection/Dataset/data.yaml',
    split='test'
)

# 3. Speed, Intensity, and Time (on Test Set)
print("\n--- [IMAGE-BY-IMAGE INTENSITY & SPEED] ---")
start_time = time.time()

# Run prediction to count "Intensity" (how many people in each frame)
results = model.predict(
    source='/kaggle/input/intrusion-detection/Intrusion Detection/Dataset/test/images',
    conf=0.25,
    save=True
)

total_time = time.time() - start_time
total_images = len(results)

for i, result in enumerate(results):
    # Intensity = Number of boxes detected in this image
    intensity = len(result.boxes)
    img_name = result.path.split('/')[-1]
    
    # Speed per image in milliseconds
    # result.speed contains: {'preprocess', 'inference', 'postprocess'}
    inf_speed = result.speed['inference']
    
    print(f"Image {i+1}: {img_name} | Intensity: {intensity} person(s) | Speed: {inf_speed:.2f}ms")

# 4. Final FPS and Summary
avg_inference = test_metrics.speed['inference']
fps = 1000 / avg_inference

print(f"\n--- [FINAL SUMMARY] ---")
print(f"Total Test Images: {total_images}")
print(f"Average Inference: {avg_inference:.2f} ms")
print(f"System Throughput: {fps:.2f} FPS 🚀")
print(f"mAP50 (Accuracy): {test_metrics.box.map50:.4f}")
```

```python
# Use 'test_metrics' which you defined in your Accuracy Report step
metrics = test_metrics 

print("--- FINAL TEST PERFORMANCE ---")
print(f"Total Images: {len(metrics.confusion_matrix.matrix)}")
print(f"Precision: {metrics.box.mp:.3f}")
print(f"Recall:    {metrics.box.mr:.3f}")
print(f"mAP@50:    {metrics.box.map50:.3f}")
print(f"mAP@50-95: {metrics.box.map:.3f}")

# Average Speed calculations
print(f"\n--- SPEED ANALYSIS ---")
print(f"Inference Speed: {metrics.speed['inference']:.2f}ms")
print(f"Total Throughput: {1000/metrics.speed['inference']:.2f} FPS")
```
