```python
import os
from ultralytics import YOLO

DATASET_ROOT = "/content/drive/MyDrive/Construction/dataset"
DATA_YAML_PATH = os.path.join(DATASET_ROOT, "data.yaml")

PROJECT_DIR = "/content/drive/MyDrive/Construction/yolov8_runs"

EXPERIMENT_NAME = "construction_yolov8n_v2_repaired"

CLASS_NAMES = [
    'HardHat', 'SafetyVest', 'Person', 'Excavator', 'Ladder',
    'Barricade', 'Tool', 'Vehicle', 'Gloves', 'SafetyGlasses'
]

yaml_content = f"""
path: {DATASET_ROOT}
train: train/images
val: val/images
test: test/images

nc: {len(CLASS_NAMES)}
names: {CLASS_NAMES}
"""

if not os.path.exists(DATASET_ROOT):
    print(f"⚠️ Warning: Base data path not found. Creating directory: {DATASET_ROOT}")
    os.makedirs(DATASET_ROOT, exist_ok=True)

with open(DATA_YAML_PATH, 'w') as f:
    f.write(yaml_content)

print(f"✅ data.yaml file successfully written at: {DATA_YAML_PATH}")

try:
    model = YOLO("yolov11n.pt")
    print("✅ YOLOv11n model loaded.")
except Exception as e:
    print(f"❌ Error loading YOLOv8n model. Please ensure the ultralytics library is installed. Error: {e}")
    exit()

print(f"\n🚀 Starting re-training with REPAIRED labels using config: {DATA_YAML_PATH}")
print(f"Results will be saved to: {os.path.join(PROJECT_DIR, EXPERIMENT_NAME)}")

results = model.train(
    data=DATA_YAML_PATH,
    epochs=50,
    imgsz=640,
    batch=16,
    name=EXPERIMENT_NAME,
    project=PROJECT_DIR,
    pretrained=True,
    workers=4
)

print(f"\n✅ Training initiated. Check your Drive for results in the '{EXPERIMENT_NAME}' folder.")
```

```python
import os
from ultralytics import YOLO

# --- Configuration ---
# This path points to the 'best.pt' file generated by your completed training run.
MODEL_PATH = "/content/drive/MyDrive/Construction/yolov8_runs/construction_yolov8n_v2_repaired3/weights/best.pt"

# This is the path to your data configuration file.
DATA_YAML_PATH = "/content/drive/MyDrive/Construction/dataset/data.yaml"

# --- Main Logic ---
if not os.path.exists(MODEL_PATH):
    print(f"❌ Error: Model file not found at: {MODEL_PATH}")
    print("Please ensure the training run completed successfully and the path is correct.")
elif not os.path.exists(DATA_YAML_PATH):
    print(f"❌ Error: Data YAML file not found at: {DATA_YAML_PATH}")
    print("Please ensure the data.yaml path is correct.")
else:
    print(f"✅ Loading model from: {MODEL_PATH}")

    try:
        # Load the newly trained custom model
        model = YOLO(MODEL_PATH)

        print(f"🚀 Starting validation using configuration from: {DATA_YAML_PATH}")

        # Run the validation mode
        # The results will be saved to the last run directory's 'val' folder.
        results = model.val(
            data=DATA_YAML_PATH,  # Path to the dataset configuration
            imgsz=640,           # Image size (same as training)
            batch=16,            # Batch size for validation
            split='val',         # Specify to use the validation split
            device=None,         # Use the default device (GPU if available)
            verbose=True,        # Print detailed per-class results
        )

        print("\n--- Validation Complete ---")
        print(f"mAP50-95: {results.box.map}")
        print(f"mAP50: {results.box.map50}")

    except Exception as e:
        print(f"❌ An error occurred during validation: {e}")
```

```python
import os
from ultralytics import YOLO

MODEL_PATH = "/content/drive/MyDrive/Construction/yolov8_runs/construction_yolov8n_v2_repaired3/weights/best.pt"
VIDEO_PATH = "/content/indianworkers.mp4"
PROJECT_DIR = "/content/drive/MyDrive/Construction/yolov8_runs/video_predictions_speed_test"
EXPERIMENT_NAME = "latency_test_indianworkers_avg_fps"

if not os.path.exists(MODEL_PATH):
    print(f"❌ Error: Model file not found at: {MODEL_PATH}")
    print("Please ensure the training run completed successfully and the path is correct.")
elif not os.path.exists(VIDEO_PATH):
    print(f"❌ Error: Video file not found at: {VIDEO_PATH}")
    print("Please ensure the video is uploaded to the /content/ folder.")
else:
    print(f"✅ Loading model from: {MODEL_PATH}")

    try:
        model = YOLO(MODEL_PATH)

        print(f"🚀 Starting AVERAGE FPS analysis on video: {VIDEO_PATH}")

        results = model.predict(
            source=VIDEO_PATH,
            conf=0.40,
            iou=0.7,
            save=False,
            project=PROJECT_DIR,
            name=EXPERIMENT_NAME
        )

        r = results[0]
        speed = r.speed

        inference_ms = speed.get('inference', 0)
        total_ms = sum(speed.values())

        print("\n" + "="*55)
        print("         ⚡ AVERAGE FPS VIDEO SPEED ANALYSIS ⚡")
        print("="*55)

        print("NOTE: Annotated video was NOT saved to maximize throughput.")
        print("-" * 55)

        print("Average Latency Metrics (ms, per image):")
        for key, value in speed.items():
            print(f"- {key.capitalize():<12}: {value:.2f} ms")

        if inference_ms > 0:
            core_fps = 1000 / inference_ms
            print("\n----------------------------------------------------")
            print(f"**CORE INFERENCE LATENCY:** {inference_ms:.2f} ms")
            print(f"**AVERAGE MODEL FPS (Inference Only):** {core_fps:.2f} FPS")

        if total_ms > 0:
            pipeline_fps = 1000 / total_ms
            print("\n----------------------------------------------------")
            print(f"**TOTAL PIPELINE LATENCY:** {total_ms:.2f} ms")
            print(f"**AVERAGE END-TO-END FPS (No Saving):** {pipeline_fps:.2f} FPS")
            print("----------------------------------------------------")

    except Exception as e:
        print(f"❌ An error occurred during prediction and analysis: {e}")
```
